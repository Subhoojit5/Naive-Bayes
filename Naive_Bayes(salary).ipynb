```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
for i in string_columns:
        df_train[i]= number.fit_transform(df_train[i])
        df_test[i]=number.fit_transform(df_test[i])
```

```python
colnames = df_train.columns
```

**Splitting the dataset in train and test for analysis**

```python
x_train = df_train[colnames[0:13]].values
y_train = df_train[colnames[13]].values
x_test = df_test[colnames[0:13]].values
y_test = df_test[colnames[13]].values
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Normalising the values**

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test =  norm_func(x_test)
```

**Applying Naive Bayes for Classification**

1. **Gaussian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB as GB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
gnb=GB()

train_pred_gau=gnb.fit(x_train,y_train).predict(x_train)
test_pred_gau=gnb.fit(x_train,y_train).predict(x_test)
```

```python
train_acc_gau=np.mean(train_pred_gau==y_train)
train_acc_gau
```

```python
test_acc_gau=np.mean(test_pred_gau==y_test)
test_acc_gau
```

```python
accu_gau=accuracy_score(y_test,test_pred_gau)
accu_gau
```

```python
confusion_matrix = confusion_matrix(y_test, test_pred_gau)
confusion_matrix
```

2. **Multinomial Naive bayes**

```python
from sklearn.naive_bayes import MultinomialNB as MB
```

```python
mnb=MB()
train_pred_multi=mnb.fit(x_train,y_train).predict(x_train)
test_pred_multi=mnb.fit(x_train,y_train).predict(x_test)
```

```python
train_acc_multi=np.mean(train_pred_multi==y_train)
train_acc_multi
```

```python
test_acc_multi=np.mean(test_pred_multi==y_test)
test_acc_multi
```

```python
accu_multi=accuracy_score(y_test,test_pred_multi)
accu_multi
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, test_pred_multi)
confusion_matrix
```

**Consolidating**

```python
t={'Model':['Gussian NB', 'Multinominal NB'], 'Accuracy':[accu_gau, accu_multi]}
T=pd.DataFrame(t)
T
```

**From the above models, it is clear that the Gaussian Naive Bayes gives a higher accuracy in classifying the salary dataset.**
